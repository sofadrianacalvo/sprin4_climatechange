```python
drop_duplicated_tweets = df.drop_duplicates(subset = ['replies_count'])
drop_duplicated_tweets
```

```python
replies_count = drop_duplicated_tweets[['tweet', 'replies_count', 'likes_count', 'retweets_count']]
replies_count.sort_values(by = 'replies_count', ascending = False).head(20)
```
